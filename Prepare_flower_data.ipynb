## Split the data to (train, validation and test)

First, let's unzip the data.

```python
# Unzip the downloaded file
import zipfile

zip_ref = zipfile.ZipFile("flowers.zip")
zip_ref.extractall()
zip_ref.close()
```

Let's see How many images we have in each folder?

```python
# How many images in each folder?
import os

# Walk through data directory and list number of files
for dirpath, dirnames, filenames in os.walk("flowers"):
  print(f"There are {len(dirnames)} directories and {len(filenames)} images in'{dirpath}'.")
```

Let's install the splitfolders 

```python
# spilt data
!pip install split-folders
```

Now, Let's spilt our data to (train, validation and test) directories

```python
# spilt data to train/val/test
import splitfolders # or import splitfolders
input_folder = "/content/flowers"
output = "/content/Flowers" #where you want the split datasets saved. one will be created if it does not exist or none is set

splitfolders.ratio(input_folder, output=output, seed=42, ratio=(.8, .15, .05)) # ratio of split are in order of train/val/test. You can change to whatever you want. For train/val sets only, you could do .75, .25 for example.
```

Let's see How many images we have in each directory?

```python
# Walk through new data directory and list number of files
for dirpath, dirnames, filenames in os.walk("Flowers"):
  print(f"There are {len(dirnames)} directories and {len(filenames)} images in'{dirpath}'.")
```
